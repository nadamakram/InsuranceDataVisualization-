# Description 
We will explore a dataset of insurance charges, to see if we can understand why some customers pay more than others.

### Import libraries 

```python
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns
```

### Load the data 

```python
data_path = r"E:\DataScienceProjects\DataVisualization\insurance.csv"
insurance_data = pd.read_csv(data_path)
insurance_data.head()
```

### Plot the data
We want to figure out the relationship or correlation between BMI(Body Mass Index) and charges

```python
sns.scatterplot(x=insurance_data['bmi'], y=insurance_data['charges'])
```

it turns out that it's a positive correlation when the bmi of a person increases, he/she most likely pays more. it's also reasonable since high BMI is typically associated with a higher risk of chronic disease.

We want to make sure of this correlation, so we'll plot a regression line.

```python
sns.regplot(x=insurance_data['bmi'], y=insurance_data['charges'])
```

Here's another correlation between charges and smoker using the bar plot, and it turns out if the person is a smoker the chance of paying more is higher than a non-smoker 

```python
sns.barplot(x=insurance_data['smoker'], y=insurance_data['charges'])
```

Now let's have a scatterplot to show a relationship between 3 factors; charges, BMI and smoker!

```python
sns.scatterplot(x=insurance_data['bmi'], y=insurance_data['charges'], hue=insurance_data['smoker'])
```

We want to make sure of this relationship, we used regression line before to plot one line, now we need 2 lines for smokers and non-smokers

```python
sns.lmplot(x="bmi", y="charges", hue="smoker", data=insurance_data)
```

Below we can use a scatterplot to describe or know a realtionship of a categorical variable like 'smoker'

```python
sns.swarmplot(x = insurance_data['smoker'], y = insurance_data['charges'])
```

we can notice that who pays more is smoker and non-smokers pay less
